```python
import sqlite3
import pandas as pd
```

```python
def load_modern_nba_data(db_path):
    # Establish a connection to the SQLite database
    connection = sqlite3.connect(db_path)
    
    # 
    query = "SELECT * FROM game WHERE season_id >= '22015'"
    
    # Use pandas to execute the query and load the data into a DataFrame
    df = pd.read_sql_query(query, connection)
    
    connection.close()
    return df

nba_df = load_modern_nba_data('nba.sqlite')

# Print the shape to confirm it worked
print(f"Success! Data Shape: {nba_df.shape}")
```
